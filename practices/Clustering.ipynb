```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import scipy.cluster.hierarchy as h
```

```python
from sklearn.datasets import load_iris
```

```python
X = load_iris(return_X_y = False)
```

```python
X
```

```python
iris = pd.DataFrame(X['data'], columns=X['feature_names'])
```

```python
iris
```

```python
iris_full = iris.copy()
iris_full['class'] = X['target']
```

```python
iris_full
```

```python
plt.figure(figsize=(15,10))
sns.pairplot(iris)
plt.show()
```

## Hierarchical

```python
Z = h.linkage(iris, method='ward', metric='euclidean')
labels = h.fcluster(Z, t=3, criterion='maxclust')
plt.figure(figsize=(15,10))
_=h.dendrogram(Z)
```

```python
iris_pred = iris.copy()
iris_pred['class'] = labels
plt.figure(figsize=(15,10))
sns.pairplot(iris_pred, hue='class')
plt.show()
```

```python
labels
```

```python
from sklearn.metrics import adjusted_mutual_info_score, silhouette_score
```

```python
adjusted_mutual_info_score(X['target'], labels)
```

```python
silhouette_score(iris, labels)
```

## K-Means

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, n_init=100)
kmeans.fit(iris)
labels = kmeans.predict(iris)
```

```python
labels
```

```python
adjusted_mutual_info_score(X['target'], labels)
```

```python
silhouette_score(iris, labels)
```

```python
kmeans.inertia_
```

## Parameter Tuning

```python
plt.figure(figsize=(15, 10))
inertia = []
for i in range(2, 30):
    kmeans = KMeans(n_clusters=i, n_init=10)
    kmeans.fit(iris)
    labels = kmeans.predict(iris)
    inertia.append(silhouette_score(iris, labels))
plt.plot(range(2, 30), inertia, 'x-r')
plt.show()
```
